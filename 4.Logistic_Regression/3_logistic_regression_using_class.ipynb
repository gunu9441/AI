# Logistic Regression with pytorch
This code is all about logistic regression which can classify two kinds of groups by using class inheriting torch.nn.module.

I focused on the structure of this code and try to memorize this sample code below.

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
```

```python
torch.manual_seed(1)
```

###Data

```python
x_data = [[1, 2], [2, 3], [3, 1], [4, 3], [5, 3], [6, 2]]
y_data = [[0], [0], [0], [1], [1], [1]]
x_train = torch.FloatTensor(x_data)
y_train = torch.FloatTensor(y_data)
```

###Model

```python
class BinaryClassifier(nn.Module):
    def __init__(self):
        super().__init__()
        self.linear = nn.Linear(2,1)
        self.sigmoid = nn.Sigmoid()

    def forward(self,x):
        return self.sigmoid(self.linear(x))
        
model = BinaryClassifier()
```

###Optimizer

```python
optimizer = optim.SGD(model.parameters(),lr=1)
```

###Training

```python
nb_epochs = 1000
for epoch in range(nb_epochs + 1):
    hypothesis = model(x_train)
    cost = F.binary_cross_entropy(hypothesis,y_train)

    optimizer.zero_grad()
    cost.backward()
    optimizer.step()

    if epoch % 100 ==0:
        prediction = hypothesis >= 0.5
        #print(type(hypothesis))
        #print(type(prediction))
        #print(type(y_train))
        correct_prediction = prediction.float() == y_train
        accuracy = correct_prediction.float().sum() / len(correct_prediction)
        print('Epoch: {:4d}/{} Cost: {:.6f} Accuracy: {:2.2f}%'.format(epoch, nb_epochs, cost.item(), accuracy*100))
```

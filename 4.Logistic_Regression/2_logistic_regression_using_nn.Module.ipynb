# Logistic Regression with pytorch
This code is all about logistic regression which can classify two kinds of groups by using torch.nn.F and nn.sequential().

I focused on the structure of this code and try to memorize this sample code below.

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
```

```python
torch.manual_seed(1)
```

###Data

```python
x_data = [[1, 2], [2, 3], [3, 1], [4, 3], [5, 3], [6, 2]]
y_data = [[0], [0], [0], [1], [1], [1]]
x_train = torch.FloatTensor(x_data)
y_train = torch.FloatTensor(y_data)
```

###Model

```python
model = nn.Sequential(
    nn.Linear(2,1),
    nn.Sigmoid()
)
```

###Optimizer

```python
optimizer = optim.SGD(model.parameters(), lr=1)
```

###Training

```python
nb_epochs = 1000
for epoch in range(nb_epochs + 1):
    hypothesis = model(x_train)
    cost = F.binary_cross_entropy(hypothesis, y_train)

    optimizer.zero_grad()
    cost.backward()
    optimizer.step()

    if epoch % 10 == 0:
        prediction = hypothesis >= torch.FloatTensor([0.5])
        correct_prediction = prediction.float() == y_train
        print(prediction.float())
        accuracy = correct_prediction.sum().item()/len(correct_prediction)
        print('Epoch: {:4d}/{} Cost: {:.6f} Accuracy {:2.2f}%'.format(epoch, nb_epochs, cost.item(), accuracy * 100))
```

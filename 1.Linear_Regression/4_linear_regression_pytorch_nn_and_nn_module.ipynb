# Linear Regression with pytorch using torch.nn and making model with torch.nn.module
This code is all about linear regression which is formed like a linear euqation graph using torch.nn library. Also, its model uses torch.nn.module which help to make a model easily.

I focused on the structure of this code and try to memorize this sample code below.

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
```

To see same results at every time, use torch.manual_seed(num)

```python
torch.manual_seed(1)
```

```python
x_train = torch.FloatTensor([[1],[2],[3]])
y_train = torch.FloatTensor([[2],[4],[6]])
```

### Model Initilization

Make model by Inheriting nn.Module

```python
class LinearRegressionModel(nn.Module):
    def __init__(self):
        super().__init__()
        self.linear = nn.Linear(1,1)
    def forward(self,x):
        return self.linear(x)
```

```python
model = LinearRegressionModel()
```

### Optimizer

```python
optimizer = optim.SGD(model.parameters(),lr=1e-2)
```

### Training

```python
nb_epochs = 2000
for epoch in range(nb_epochs+1):
    prediction = model(x_train)
    cost = F.mse_loss(prediction, y_train)

    optimizer.zero_grad()
    cost.backward()
    optimizer.step()

    if epoch % 100 == 0:
        print("Epoch {:4d}/{} Cost: {:.6f}".format(epoch, nb_epochs, cost.item()))
```

# Multivariable Linear Regression with pytorch using torch.nn
This code is all about Multivariable linear regression which is formed like a linear euqation graph using torch.nn library and torch.nn.module.

I focused on the structure of this code and try to memorize this sample code below.

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
```

To see same results at every time, use torch.manual_seed(num)

```python
torch.manual_seed(1)
```

### Data

```python
x_train = torch.FloatTensor([[73, 80, 75],
                             [93, 88, 93],
                             [89, 91, 90],
                             [96, 98, 100],
                             [73, 66, 70]])
y_train = torch.FloatTensor([[152], [185], [180], [196], [142]])
```

### Model

```python
class MultivariateLinearRegressionModel(nn.Module):
    def __init__(self):
        super().__init__()
        self.linear = nn.Linear(3,1)
    
    def forward(self,x):
        return self.linear(x)
```

```python
model = MultivariateLinearRegressionModel()
```

### Optimizer

```python
optimizer = optim.SGD(model.parameters(),lr = 1e-5)
```

### Training

```python
nb_epochs = 2000
for epoch in range(nb_epochs + 1):
    prediction = model(x_train)
    cost = F.mse_loss(prediction, y_train)

    optimizer.zero_grad()
    cost.backward()
    optimizer.step()

    if epoch%100 == 0 :
        print('Epoch: {:4d}/{} Cost:{:4.6f}'.format(epoch,nb_epochs,cost.item()) )
```

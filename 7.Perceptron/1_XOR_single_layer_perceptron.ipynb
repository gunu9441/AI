# XOR model by using single-layer perceptron
This code below is all about solving XOR problem  by using single-layer perceptron,which can not solve this problem. It is for demonstrating a single-layer perceptron code simulation.

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader
from torch.utils.data import Dataset
```

```python
device = 'cuda' if torch.cuda.is_available() else 'cpu'
print('device: ', device)
```

In general, dataset class get numpy array and return torch.Tensor which is converted from ndarray.

```python
class xor_dataset(Dataset):
    def __init__(self):
        #self.x_train = torch.FloatTensor([[0, 0], [0, 1], [1, 0], [1, 1]]).to(device)
        #self.y_train = torch.FloatTensor([[0], [1], [1], [0]]).to(device)
        self.x_train = [[0, 0], [0, 1], [1, 0], [1, 1]]
        self.y_train = [[0], [1], [1], [0]]
    def __len__(self):
        return len(self.x_train)
    def __getitem__(self, idx):
        #return self.x_train[dix], self.y_train[idx]
        return torch.FloatTensor(self.x_train[idx]).to(device), torch.FloatTensor(self.y_train[idx]).to(device)

dataset = xor_dataset()
```

```python
dataloader = DataLoader(dataset=dataset, batch_size = 2, shuffle=True, drop_last=False)
iteration = len(dataloader)
```

```python
class multi_layer_model(nn.Module):
    def __init__(self):
        super().__init__()
        self.layer = nn.Sequential(
            nn.Linear(2,1),
            nn.Sigmoid()
        ).to(device)
    def forward(self, x):
        out = self.layer(x)
        return out
```

```python
model = multi_layer_model()
```

```python
cost_function = nn.BCELoss().to(device)
```

```python
optimizer = optim.SGD(model.parameters(),lr = 1e-3)
```

```python
nb_epochs = 10000

for epoch in range(nb_epochs+1):
    for idx, (X,Y) in enumerate(dataloader):
        prediction = model(X)
        #print(prediction)
        #print(Y)
        cost = cost_function(prediction, Y)

        optimizer.zero_grad()
        cost.backward()
        optimizer.step()

        avg_cost = cost / iteration
        print('idx: {:3d}/{}'.format(idx+1, iteration))
    print('Epoch: {:4d}/{}, Cost: {:.5f}'.format(epoch, nb_epochs, cost.item()))
```

```python
with torch.no_grad():
    X = torch.FloatTensor([[0, 0], [0, 1], [1, 0], [1, 1]]).to(device)
    Y = torch.FloatTensor([[0], [1], [1], [0]]).to(device)
    hypothesis = model(X)
    prediction = (hypothesis > 0.5)
    accuracy = (prediction == Y).float().mean()
    print('Hypothesis: ',hypothesis.detach().cpu().numpy())
    print('Prediction: ',prediction.detach().cpu().numpy())
    print('Y: ', Y.detach().cpu().numpy())
    print('Accuracy: ', accuracy.item())
```
